```python
import igraph as ig
```

```python
g = ig.Graph.Read('../data/processed/trenord.graphml', format='graphml')
```

```python
for e in g.es.select(lambda e: e.source_vertex['label']=='BUSTO ARSIZIO FN' and e.target_vertex['label']=='CASTELLANZA'):
    print(e)
```

```python
es_r27 = { e for e in g.es if e['label'] == 'R27' }
g_r27 = g.subgraph_edges(es_r27)
```

```python
visual_style = {} #Python dictionary 

ig.plot(g_r27, bbox=(0, 0, 1000, 1000), **visual_style)
```
